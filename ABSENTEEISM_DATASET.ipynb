## Absenteeism Analysis Project ##

This project analyzes employee absenteeism data to uncover patterns and potential causes of absence in the workplace. By examining various features such as reasons for absence, health status, body mass index (BMI), and work habits, we aim to build insights that could help organizations improve productivity and employee well-being.

The analysis includes:
- Data loading and exploration
- Cleaning and preprocessing
- Feature engineering
- Logistic regression modeling
- Evaluation of results

Dataset Source: `Absenteesim_Data.csv`

## Extracting Day of the Week

```python
df_reason_mod["Date"][0].weekday()
```

```python
def day_to_week(date_values):
    return date_values.weekday()
```

```python
df_reason_mod["Day of the Week"] = df_reason_mod["Date"].apply(day_to_week)
```
